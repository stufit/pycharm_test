```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import os
import re
import json

url = 'https://tv.naver.com/v/16389068/list/67096'
        

res = requests.get(url)
print(res.status_code)

if res.ok:
    soup=BeautifulSoup(res.text,'lxml')
    print(soup)
    div_tag = soup.select('div p')[0].text
#     print(p_tag)
    json_data = json.loads(div_tag)
    print(type(json_data), json_data)
    
    episode_length = len(json_data['data'])
    
    print(len(json_data['data']),type(json_data['data']))
    episode_list = json_data['data']  
    
    for episode in episode_list:
        print('------')
        #에피소드의 제목
        title = episode['title']
        #에피소드 다운로드 url
        mp4_url = episode['enclosure']['url']
        print(title, mp4_url)
        
        if mp4_url:
            
            req_header = {
                'referer':'https://tv.naver.com/v/16389175',
                'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.111 Safari/537.36'
            }
            
            dir_name = os.path.join('naver','야구')
            print(dir_name) #mp3\여바라
            
            #해당 디렉토리가 없으면
            if not os.path.isdir(dir_name):
                #디렉토리를 생성
                os.makedirs(dir_name)
            
            #해당 url에서 파일명만 추출
            file_name = os.path.basename(mp4_url)
            #생성된 디렉토리명과 파일명을 합쳐주기
            file_name = dir_name+'/'+file_name
            
            #print(file_name)
            res = requests.get(mp4_url, headers=req_header)
            if res.ok:
                #response 객체에서 binary 데이터 추출
                mp4_bin = res.content                
                #binary 데이터를 local 파일로 저장
                with open(file_name, 'wb') as file:
                    file.write(mp3_bin)
```

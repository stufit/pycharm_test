### 3. Json 파일의 내용을 MongoDB에 저장하기

```python
import pymongo

con = pymongo.MongoClient()
db = con.member_db
col = db.members
print(col)
```

```python
col.insert_many(member_json)
print(col.count_documents({}))
```

```python
# 정당이 열린민주당인 국회의원은?
def print_member(criteria,projection):
    for idx,member in enumerate(col.find(criteria,projection),1):
        print(idx,member)
    

criteria = {'정당':'열린민주당'}
projection = {'_id':0,'이름':1,'정당':1,'선거구':1}
print_member(criteria,projection)
```

```python
#컬렉션의 aggregate()를 사용한 함수 정의
def print_aggregate_document(*args):
    agg_list=[]
    for arg in args:
        agg_list.append(arg)
    print(agg_list)
    
    for agg_doc in col.aggregate(agg_list):
        print(agg_doc)
```

```python
# 1.정당별 국회의원 몇명인지?
group={
    '$group':{'_id':'$정당','정당별개수':{'$sum':1}}
}
project={
    '$project':{'정당':'$_id','_id':0,'정당별개수':1}
}
sort={
    '$sort':{'정당별개수':-1}
}
print_aggregate_document(group,project,sort)
```

```python
# 1-2.정당별 갯수
for member in col.aggregate([
    {'$group':{'_id':'$정당','정당개수':{'$sum':1}}},
    {'$sort':{'정당개수':-1}},
    {'$project':{'_id':0,'정당':'$_id','정당개수':1}}
    
]):
    print(member)
```

```python
# 지역(area)별 국회의원은 몇명인지? 많은 순서대로
# 지역별개수:10, 지역:'경기'
group={
    '$group':{'_id':'$선거구','선거구별개수':{'$sum':1}}
}
print_aggregate_document(group)
```

```python
# 취미가 등산인 국회의원은?
for member in col.find({'취미, 특기':'등산'},{'_id':0,'이름':1}):
    print(member)
```

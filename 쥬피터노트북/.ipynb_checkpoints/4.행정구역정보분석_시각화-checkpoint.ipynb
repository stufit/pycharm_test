```python
#강원도에서 가장 높은 인구수는?
pop_max_value = data_df.loc[data_df['광역시도']=='강원도','인구수'].max()
#print(pop_max_value)
data_df.loc[data_df['인구수']==pop_max_value,['광역시도','행정구역','인구수']] 
```

```python
#위의 것 함수로 만들어서 광역시도 중에서 인구수가 가장 많은 행정구역 반환
def get_pop_max_value(sido):
    pop_max_value = data_df.loc[data_df['광역시도']== sido,'인구수'].max()
    return data_df.loc[data_df['인구수']==pop_max_value,['광역시도','행정구역','인구수']] 

get_pop_max_value('서울특별시')
```

```python
for sido_name in data_df['광역시도'].unique():
   # print(sido_name)
     print(get_pop_max_value(sido_name))
```

```python
# 새로운 DataFrame 객체를 생성하기

pop_max_df = pd.DataFrame(columns=['광역시도','행정구역','인구수'])

for sido_name in data_df['광역시도'].unique():
    pop_max_df = pop_max_df.append(get_pop_max_value(sido_name))
    
pop_max_df = pop_max_df.reset_index(drop=True)
pop_max_df
```

```python
# 0부터 시작하는 index를 1부터 시작하도록 바꾸기
# index 확인
pop_max_df.index
```

### 광역시도별 면적이 가장 넓은 지역

* 광역시도, 행정구역, 면적 컬럼
* 면적이 가장 넓은 순서대로 정렬
* index는 1부터 시작하도록
* 최종 DataFrame을 excel파일로 저장하기


```python
# 면적이 가장 넓은 광역시도 및 행정구역
pop_max_men = data_df.loc[data_df['광역시도']=='강원도','면적'].max()
print(pop_max_men)
data_df.loc[data_df['면적']==pop_max_men,['광역시도','행정구역','면적']] 

# 면적이 가장 넓은 광역시도 및 행정구역들을 나열
def get_pop_max_men(sidoo):
    pop_max_men = data_df.loc[data_df['광역시도']==sidoo,'면적'].max()
    return data_df.loc[data_df['면적']==pop_max_men,['광역시도','행정구역','면적']].sort_values(by='면적',ascending=True)


for sidoo_name in data_df['광역시도'].unique():
    print(get_pop_max_men(sidoo_name))
```

```python
# 컬럼 테이블로 만들기
pop_max_men_df = pd.DataFrame(columns=['광역시도','행정구역','면적'])
for sidoo_name in data_df['광역시도'].unique():
    pop_max_men_df = pop_max_men_df.append(get_pop_max_men(sidoo_name))
    
pop_max_men_df = pop_max_men_df.sort_values(by='면적',ascending=False).reset_index(drop=True)
pop_max_men_df
    
```

```python
# 인덱스 1부터 시작하게 만들기
pop_max_men_df.index = np.arange(1,len(pop_max_men_df)+1)

pop_max_men_df
```

```python
# 엑셀 파일로 저장
pop_max_men_df.to_excel('data/광역시도별면적최대지역.xlsx',sheet_name='최대면적지역들')
```

### 상관계수
* 인구수와 면적 간의 어느 정도의 관련성이 있는지 보기 위해서 상관계수 구하기
* 상관계수는 -1부터 +1 사이의 값
* 0에 가까울수록 서로 관련성이 없으며, -1 or 1에 가까울수록 서로 관련성이 있다.
* corr( ) 함수를 사용


```python
# 전체 행정구역에서의 면적과 인구수의 상관계수를 구하시오

data_df['면적'].corr(data_df['인구수'])
```

```python
def get_corr_value(sido):
    sido_df = data_df.loc[data_df['광역시도']==sido]
    return sido_df['면적'].corr(sido_df['인구수'])
```

```python
get_corr_value('서울특별시')
```

### Group By 기능 사용하기
* Series 객체의 groupby( ) => 컬럼값을 인자로 전달한다.
* DataFrame 객체의 groupby( ) => 컬럼명을 인자로 전달한다.
* ~별(ex]광역시도별)에 해당하는 컬럼명은 groupby( ) 함수의 인자로 전달한다.

```python
#광역시도별 인구수의 합계 구하기
data['인구수'].groupby(data['광역시도']).sum()
```

```python
#grouping 한 결과를 정렬
data['인구수'].groupby(data['광역시도']).sum().sort_values(ascending=False)
```

```python
#광역시도별 인구수의 합계 - DataFrame 객체
#data['인구수'].groupby(data['광역시도'])

#DataFrameGroupBy 객체를 반환-> columns를 반환하는 것
data.groupby('광역시도').sum()
```

```python
#DataFrameGroupBy 인구수의 컬럼 반환
data.groupby('광역시도')['인구수'].sum()
```

```python
# 광역시도별 면적의 합계를 구하고, 넓은 순서대로 정렬하시오.
data.groupby('광역시도')['면적'].sum().sort_values(ascending=False)
```

```python
# 광역시도별, 행정구역별 인구수의 합계
group_df = data.groupby(['광역시도','행정구역'])['인구수'].sum()
# 엑셀 저장
group_df.to_excel('data/광역시도별행정구역별인구수.xlsx',sheet_name='인구수')
```

```python
# 광역시도별 인구수의 max, mean, std 구하기
# 여러개의 집계함수를 한꺼번에 사용하려면 agg() 함수를 사용하면 됨.
data.groupby('광역시도')['인구수'].agg(['max','mean','std'])
```

```python
#위의 결과값 중에서 세종특별자치시의 std 가 NaN인 것을 숫자 0으로 바꾸기
# numpy의 std() 함수를 호출한 결과를 agg() 인자로 전달
import numpy as np

def std(x):
    return np.std(x)

data.groupby(['광역시도'])['인구수'].agg(['max','min',std]).sort_values(by='std',ascending=False)
```

```python
# 인덱스를 컬럼으로 바꾸는 작업( index에 있는 광역시도를 columns 쪽으로 옮기는 작업)
# 변환 시 reset_index()함수의 level 속성에 변환할 컬럼명을 설정
# reset_index()의 inplace의 기본값(default)이 False => 객체를 변경하지는 않고 출력만 함.
# inplace의 값이 True 이면 객체를 변경하고, 출력은 하지 않는다.
group_agg_df = data.groupby(['광역시도'])['인구수'].agg(['max','min',std]).sort_values(by='std',ascending=False)

group_agg_df.reset_index(level='광역시도', inplace=True)

```

```python
# 인덱스 1부터 시작하게 만들기
group_agg_df.index = np.arange(1,len(group_agg_df)+1)

group_agg_df
```

```python
# 행과 열을 뒤집어 주어서 출력하는 것
group_agg_df.T
# .T = .transpose()
```

```python
group_agg_df.transpose()
```
